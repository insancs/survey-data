```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
survey_data = pd.read_csv('E:/DQLab/Dataset/Datacamp/young-people-survey-responses.csv', sep=';')

# Display the first n rows for the object based on position. Default 5 rows
survey_data.head()
```

```python
# Display dataframe information
survey_data.info()
```

```python
# Check number of missing value after imputation 
print(survey_data.isna().sum())
```

```python
# Imputation missing value using median 
survey_data = survey_data.fillna(survey_data.median())

# Check number of missing value after imputation 
print(survey_data.isna().sum())
```

```python
# Delete index column
survey_data = survey_data.drop(['index'], axis=1)

# Delete missing value
survey_data = survey_data.dropna(axis=0)

# Check the data type after conversion
print(survey_data.info())
```

```python
gender = survey_data['Gender'].value_counts()
label = ["Female", "Male"]

plt.pie(gender, labels = label, autopct="%1.2f%%")
plt.title("Gender Percentage", fontsize=20, pad=10)
plt.gcf().set_size_inches(5, 5)
plt.savefig('1_gen dist.png')
plt.show()
```

```python
sns.countplot(x='Age', data=survey_data)
plt.title("Age Distribution", fontsize=20, pad=15)
plt.gcf().set_size_inches(9, 5)
plt.savefig("2_age dist.png")
plt.show()
```

```python
sns.set_style("whitegrid")
# Create a count plot with "Spiders" on the x-axis
sns.countplot(x="Mathematics", data=survey_data)
plt.title("Interested in Mathematics", fontsize=20, pad=15)
plt.gcf().set_size_inches(7, 5)
plt.savefig("3_interest math.png")
plt.show()
```

```python
# Create a count plot with "Spiders" on the x-axis
sns.countplot(x="Spiders", data=survey_data)
plt.title("Phobia to Spiders", fontsize=20, pad=15)
plt.gcf().set_size_inches(7, 5)
plt.savefig("4_spiders.png")
plt.show()
```

```python
# Create column subplots based on age category
orders = ['less than an hour a day', 'few hours a day', 'most of the day', 'no time at all']
sns.catplot(x="Internet usage", data=survey_data,
            kind="count", 
            hue="Gender",
            order=orders)

plt.title("Internet Usage by Gender", fontsize=20, pad=15)
plt.gcf().set_size_inches(10, 7)
plt.savefig("5_internet usage.png")
plt.show()
```

```python
# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents' advice", 
            data=survey_data, 
            kind="count")
plt.title("Young People Listen to Parents' Advice ", fontsize=20, pad=15)
plt.gcf().set_size_inches(7, 5)
plt.savefig("6_parents.png")
plt.show()
```

```python
# Create the box plot of age distribution by gender
sns.catplot(x="Gender", y="Age", 
            data=survey_data, 
            kind="box")
plt.gcf().set_size_inches(6, 6)
plt.savefig("7_gen age.png")
plt.show()
```

```python
# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender", y="Age", 
                data=survey_data, 
                kind="box", 
                hue="Pets")

# Set title to "Age of Those Interested in Pets vs. Not"
g.fig.suptitle("Age of Those Interested in Pets vs. Not")
plt.gcf().set_size_inches(9, 5)
plt.savefig("8_gen age pets.png")
plt.show()
```

```python
# Adjust to add subplots per gender
g = sns.catplot(x="Village - town", y="Techno", 
                data=survey_data, 
                kind="bar",
                hue="Gender",
                ci=None)

# Add title and axis labels
g.fig.suptitle("Young People Who Like Techno", y=1.02, fontsize=20)
g.set(xlabel="Location of Residence", ylabel="Who Like Techno")
plt.gcf().set_size_inches(8, 6)
plt.savefig("9_loc tech.png")
plt.show()
```

```python
sns.countplot(x='Finances', 
              data=survey_data, 
              hue='Village - town')
plt.title("Resindence with Stable Finance", fontsize=20, pad=15)

plt.savefig("10_loc finance.png")
plt.show()
```
